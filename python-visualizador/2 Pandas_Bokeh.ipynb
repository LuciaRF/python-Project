### Ejercicio hacer el siguiente gráfico de dos formas:

1 con pandas    
2 con python básico (por ejemplo usando diccionarios)

Sacar también sus valores estadísticos con pandad y con python clasico

### Solución

```python
index=1
plt.figure(figsize=(20,20))
for i in df["dataset"].unique():
    plt.subplot(4,4,index)
    x=df[df["dataset"]==i]["x"]
    y=df[df["dataset"]==i]["y"]
    plt.scatter(x,y,s=10)
    plt.title("dataset: "+i)
    index+=1
```

```python
# estadísticas con Pandas:
for i in df["dataset"].unique():
    x=df[df["dataset"]==i]["x"]
    y=df[df["dataset"]==i]["y"]
    x_mean=np.mean(x)
    y_mean=np.mean(y)
    x_std=np.std(x)
    y_std=np.std(y)
    corr=np.corrcoef(x, y)[0][1]
    print("Dataset: %10s   X-Mean: %7.2f   Y-Mean: %7.2f   X_STD: %7.2f   Y_STD: %7.2f   PEARSONS_CORR: %7.2f" % (i,x_mean,y_mean,x_std,y_std,corr))
```

#### Plot Dataset usando Python básico ( no pandas)

```python
dictionary={}
with open("DatasaurusDozen.tsv","r") as f:
    header=f.readline().split("\t")
    for row in f:
        key,x,y=row.split("\t")
        if key not in dictionary:
            dictionary[key]={"x":[float(x)],"y":[float(y)]}
        else:
            dictionary[key]["x"].append(float(x))
            dictionary[key]["y"].append(float(y))
```

```python
index=1
plt.figure(figsize=(20,20))
for i in dictionary.keys():
    plt.subplot(4,4,index)
    x=dictionary[i]["x"]
    y=dictionary[i]["y"]
    plt.scatter(x,y,s=10)
    plt.title("dataset: "+i)
    index+=1
```

```python
#Estadísticas con Python Básico
for i in dictionary.keys():
    x=dictionary[i]["x"]
    y=dictionary[i]["y"]
    x_mean=np.mean(x)
    y_mean=np.mean(y)
    x_std=np.std(x)
    y_std=np.std(y)
    corr=np.corrcoef(x, y)[0][1]
    print("Dataset: %10s   X-Mean: %7.2f   Y-Mean: %7.2f   X_STD: %7.2f   Y_STD: %7.2f   PERASONS_CORR: %7.2f" % (i,x_mean,y_mean,x_std,y_std,corr))
```

### 1. Hacemos nuestro primer plot en Bokeh

```python
# preparamos unos datos artificiales
x = [0.1, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0]
y0 = [i**2 for i in x]
y1 = [10**i for i in x]
y2 = [10**(i**2) for i in x]
```

```python
#Definimos el plot:
p = figure(
   tools="pan,box_zoom,reset,save",
   y_axis_type="log", y_range=[0.001, 10**11], title="log axis example",
   x_axis_label='sections', y_axis_label='particles'
)
```

```python
#Añadimos renders
p.line(x, x, legend_label="y=x")
p.circle(x, x, legend_label="y=x", fill_color="white", size=8)
p.line(x, y0, legend_label="y=x^2", line_width=3)
p.line(x, y1, legend_label="y=10^x", line_color="red")
p.circle(x, y1, legend_label="y=10^x", fill_color="red", line_color="red", size=6)
p.line(x, y2, legend_label="y=10^x^2", line_color="orange", line_dash="4 4")
```

```python
#Mostrarremos los resultamos en un fichero html que crearemos:
# Save file
output_file("./ejemplo1bokeh.html", "Este es mi primer gráfico en HTML")
show(p)


from IPython.display import Javascript

url = 'http://127.0.0.1:8888/view/ejemplo1bokeh.html'
display(Javascript('window.open("{url}");'.format(url=url)))
```

### 2. Añadiendo herramientas a un gráfico. Prueba los nuevos iconos que te aparecen

```python
import numpy as np

from bokeh.plotting import figure, output_file
from bokeh.io import output_notebook, show
output_notebook()

# prepare some data
N = 4000
x = np.random.random(size=N) * 100
y = np.random.random(size=N) * 100
radii = np.random.random(size=N) * 1.5
colors = ["#%02x%02x%02x" % (int(r), int(g), 150) for r, g in zip(50+2*x, 30+2*y)]

TOOLS = "crosshair,pan,wheel_zoom,box_zoom,reset,box_select,lasso_select"
p = figure(tools=TOOLS, x_range=(0, 100), y_range=(0, 100))
p.circle(x, y, radius=radii, fill_color=colors, fill_alpha=0.6, line_color=None)
show(p)
```

```python
# !pip install pandas_datareader
# !pip install yfinance
import numpy as np
import pandas as pd
from pandas_datareader import data as data


import yfinance as yf
yf.pdr_override()
# Define the instruments to download. We wou
ld like to see Apple and Microsoft
tickers = ['AAPL', 'MSFT']

# We would like all available data from 01/01/2000 until 12/31/2016.
start_date = '2010-01-01'
end_date = '2019-01-27'

#OBTENEMOS LOS DATOS DE YAHOO LLAMANDO A LA API
panel_data = data.get_data_yahoo(tickers, start_date, end_date)



print(panel_data.head())
```

```python
#OJO que tiene índices multicolumna. Para verlo lo pivoto

#Visualizamos el data frame pivotado
# Stack: pivota nuestra tabla devolviéndonos un data frame con un indi# ce con un nuevo nivel más interno 
df=panel_data.stack()
df.head()
```

```python
df2=df.reset_index(drop=False)
```

```python
df2
```

```python
df2=df2.rename(columns={"level_1": "Company"})
```

```python
df2
```

```python
df3=df2[df2["Company"]=="AAPL"]
```

```python
df3
```

```python
# Getting just the adjusted closing prices.
close = panel_data['Close']

# Getting all weekdays (no weekends) between 01/01/2000
all_weekdays = pd.date_range(start=start_date, end=end_date, freq='B')

# How do we align the existing prices in adj_close with our new set of dates?
# All we need to do is reindex close using all_weekdays as the new index
close = close.reindex(all_weekdays)

# Reindexing will insert missing values (NaN) for the dates that were not present
# in the original set. To cope with this, we can fill the missing by replacing them
# with the latest available price for each instrument.
close = close.fillna(method='ffill')
close.head()
```

### 12. Mostrando informacion al pasar el ratón ( Hover)

```python
from bokeh.models import HoverTool


source = ColumnDataSource(
        data=dict(
            x=df3.index,
            y=df3["Close"],
            high=df3["High"],
            volume=df3["Volume"],
        )
    )

hover = HoverTool(
        tooltips=[
            #("index", "$index"),
            ("company", "Apple"),
            ("Date", "@x{%F}"),
            ("Close","$y"),
            ("High","@high"),
            ("Volume","@volume"),
        ],
        formatters={"x":"datetime"}
    )

p = figure(plot_width=600, plot_height=300, x_axis_type="datetime",tools=[hover], title="Mouse over the dots")
p.line('x','y', source=source)

show(p)
```
