```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph

    Parameters
    ----------
    ax : Axes
        The axes to draw to

    data1 : array
       The x data

    data2 : array
       The y data

    param_dict : dict
       Dictionary of kwargs to pass to ax.plot

    Returns
    -------
    out : list
        list of artists added
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, (ax1, ax2) = plt.subplots(1, 2)
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```

### Figuras mltiples y ejes

```python
def coseno(t):
    return np.cos(2*np.pi*t)

# Funcin cos multiplicada por una exponencial decreciente
def cosdec(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)   # Separacin de los puntos en el eje X
t2 = np.arange(0.0, 5.0, 0.04)  # Escala para la lnea

plt.figure(figsize=(12,4))
plt.figure(1)
plt.subplot(121)
plt.plot(t1, cosdec(t1), 'bo', t2, cosdec(t2), 'k')

plt.subplot(122)
plt.plot(t1, coseno(t1), 'rs', t2, coseno(t2), 'r--')
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

### Histogramas. Aadir texto y grid

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# Histograma, el segundo parmetro es el nmero de bins
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('CI')
plt.ylabel('Probabilidad')
plt.title('Distribucin del CI')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
np.random.seed(2022)

# example data
mu = 100  # mean of distribution
sigma = 15  # standard deviation of distribution
x = mu + sigma * np.random.randn(437)

num_bins = 20

fig, ax = plt.subplots()

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins, density=1)

# add a 'best fit' line applying the Gaussian formula 
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * ((bins - mu)/ sigma )**2))

ax.plot(bins, y, '--r')
ax.set_xlabel('Smarts')
ax.set_ylabel('Probability density')
ax.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
```

### Ley de los grandes nmeros

La ley de los grandes nmeros indica que la media de la suma de N variables aleatorias independientes tiende a una 
distribucin normal.

```python
# Generamos un nmero muy grande de valores de una distribucin uniforme.
# El histograma es plano
np.random.seed(2022)
nmuestras = 100000
un = np.random.rand(nmuestras)

# Histograma
n, bins, patches = plt.hist(un, 10, density=1, facecolor='b', alpha=0.2)
plt.title('Distribucin de la variable uniforme')
```

```python
# Demostracin de cmo funciona el lmite central
nmedias = 500
tsamps = 100
mean_muestras = np.zeros(nmedias)                      # Vector para almacenar las medias
for i in range(0,nmedias):
    samp = np.random.choice(un, tsamps, replace=False) # Escogemos una pequea muestra de nmeros al azar
    mean_muestras[i] = np.mean(samp)

# Histograma
n, bins, patches = plt.hist(mean_muestras, 20, density=1, facecolor='g', alpha=0.2)
plt.title('Distribucin de la media')
plt.ylabel('Densidad de probabilidad')
```

```python
# Repetir cdigo es una malsima prctica de programa. Habra que incluir el muestreo en una funcin.
plt.figure(1, figsize=(15, 4))
vmedias = [50,500,10000]
j = 1
for nmedias in vmedias:
    mean_muestras = np.zeros(nmedias)                      # Vector para almacenar las medias
    for i in range(0,nmedias):
        samp = np.random.choice(un, tsamps, replace=False) # Escogemos una pequea muestra de nmeros al azar
        mean_muestras[i] = np.mean(samp)
    plt.subplot(130+j)
    # Histograma
    n, bins, patches = plt.hist(mean_muestras, 50, density=1, facecolor='g', alpha=0.2)
    plt.title(f'Distribucin de la media: {nmedias} muestras')
    plt.ylabel('Densidad de probabilidad')
    j = j + 1
    # lnea de tendencia 
    y = ((1 / (np.sqrt(2 * np.pi) * np.std(mean_muestras))) * np.exp(-0.5 * ((bins - np.mean(mean_muestras))/ np.std(mean_muestras) )**2))
    plt.plot(bins, y, 'r--')
plt.show()
```

### Diferentes tipos de ejes a dems del lineal

```python
from matplotlib.ticker import NullFormatter  # useful for `logit` scale

# Fixing random state for reproducibility
np.random.seed(2022)

# make up some data in the interval ]0, 1[
y = np.random.normal(loc=0.5, scale=0.4, size=1000)  # np.random.normal(mean, std, size)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y)) # to generate the X axis

# plot with various axes scales

plt.figure(figsize=(8,3))

# linear
plt.subplot(121)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)


# log
plt.subplot(122)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)


# Format the minor tick labels of the y-axis to avoid too many labels.
plt.gca().yaxis.set_minor_formatter(NullFormatter())
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"

plt.subplots_adjust(top=1.2, bottom=0.08, left=0.10, right=0.95, hspace=0.25, wspace=0.35)

plt.show()
```

### Leyendas customizadas
